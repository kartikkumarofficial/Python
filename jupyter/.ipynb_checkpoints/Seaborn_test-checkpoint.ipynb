## JoinPlot
allows to study relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot , a hexbin plot , a 2d histogram , or a 2d density plot.

### Bivariate Analysis

```python
# sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
# this line at sides = probability density function (will study later)
```

## Pairplot
also known as as scatterplot , in which one variable in the same data row is matched with another variable's value like :Pair plots are elaborations showing all paired variables paired with all the other variables 

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker') 
# hue= gives on which value basis will the values be compared
```

```python
df['smoker'].value_counts()
```

## Dist Plot
Dist plot helps us to check the distribution of the columns feature

```python
sns.displot(df['tip'],kde='True') 
# kde = Kernerl Density Estimation - smoothed version of histogram showing probability density function of data 
```

```python
sns.displot(df['tip'],kde='True',bins=10)
# will divide the data in 10 equal parts
```

## Categorical Plots
helps in analysis of categorical data points (non numerical values in table)
*boxplot
*violinplot
*countplot
*barplot


```python
# Countplot
# sns.countplot(df['sex'])
sns.countplot(x='sex',data=df,palette=['g','orange'],hue='sex',legend=True)
# palette - list that defines different colors for values
# hue- added coz palette used individually is deprecated 
# legend - writes values on top
```

```python
sns.countplot(x='day',data=df,palette=['g','r','b','cyan'],hue='day',legend=False)
```

```python
# for horizontal representation
sns.countplot(y='sex',data=df,palette=['g','orange'],hue='sex')
```

## Barplot


```python
# bar plot - need to specify x axis value and y axis value
sns.barplot(x='total_bill',y='sex',data=df,legend=True,palette={'Male': 'skyblue', 'Female': 'pink'})
```

## Boxplot
a box or whisker plot is a graph that presents info from a five number summary.
quickly shows skewness


```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='total_bill',y='day',data=df,hue="smoker")
```

## Violinplot
helps to see both the distribution of data in terms of kde(kernel density estimation) as well as box plot


```python
sns.violinplot(x='total_bill',y='day',hue='smoker',data=df)
```
