```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Creating an http req using requests library
requests.get("https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page=500").text
```

```python
# content must be inside robots.txt whwich makes it unavailable for bots to access
# solution - disguise as u are a web browser accessing data lol
# we'll add headers - extra piece of info which will make the website believe that the request is not coming from a bot but rather a browser
```

```python

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/122.0.0.0 Safari/537.36",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
    "Referer": "https://www.ambitionbox.com/",
    "Connection": "keep-alive"
}
webpage = requests.get("https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page=500",headers=headers).text
```

```python
# BeautifulSoup is a library used for web scraping
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
# print(soup.prettify())
```

```python
soup.find_all('h1')[0].text
```

```python
for i in soup.find_all('h2'):
    print(i.text.strip())
```

```python
for j in soup.find_all('div',class_='rating_text rating_text--md'):
    print(j.text.strip())
```

```python
# similarly for other attributes :
# 1-inspect elements on webpage 
# 2-choose the tag and tag associated and then use it here in using BeautifulSoup Library
# 3-add the data in an empty list
```

```python
company=soup.find_all('div', class_="companyCardWrapper__metaInformation")
company
```

```python
name=[]
rating=[]
# other params
```

```python
for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('div',class_='rating_text rating_text--md').text.strip())
```

```python
d={'name':name , 'rating':rating}
```

```python
df=pd.DataFrame(d)
```
