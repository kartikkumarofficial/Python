```python
# Independent and Dependent Variables
X=dataset.drop("MEDV",axis=1) # take all columns of dataset except medv
y=df["MEDV"] # take medv column as dependent variable
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
# standardising the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit_transform(X_train)
```

```python
X_train=scaler.fit_transform(X_train)
# scaler.fit() + scaler.transform() - to directly make changes 
```

```python
X_test = scaler.transform(X_test)
```

```python
# to revert changes
# scaler.inverse_transform(x_train)
```

```python
# implementing linear regression
from sklearn.linear_model import LinearRegression
# cross validation
# Cross-validation is a technique to evaluate your model more reliably by using the dataset multiple times 
# for both training and testing.
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(X_train,y_train)
```

```python
# to find scoring => https://scikit-learn.org/stable/modules/model_evaluation.html
mse=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=10)
mse
```

```python
np.mean(mse)
```

```python
# prediction
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

```python
# to verify kitna sahi hai , we will verify these values with truth value i.e.y_test
```

```python
sns.displot(reg_pred-y_test,kind='kde')
```

```python
# this graph shows most data ka deviation 10 ki range me hi hai + and - so the data
# is good , only outliers ka zyada hai => -20 , -30
# this shows the model has made good predictions
```

```python
from sklearn.metrics import r2_score
score = r2_score(reg_pred,y_test)
```
