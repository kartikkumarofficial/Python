```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/files/programming/Python/2025/jupyter/datasets/titanic_mod.csv")
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig=df['number'].value_counts().plot.bar()
```

```python
# extract numerical part
df['number_numerical'] = pd.to_numeric(df['number'],errors='coerce',downcast = 'integer')
```

```python
# extract categorical part
df['number_categorical'] = np.where(df['number_numerical'].isnull(),df['number'],np.nan)
```

```python
df.head()
```

```python
# handling cabin
```

```python
# capturing numeric part 
df['cabin_num']=df['Cabin'].str.extract('(\d+)')
```

```python
# capturing f irst letter 
df['cabin_cat'] = df['Cabin'].str[0]
```

```python
df.head()
```

```python
# extracting last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s:s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],errors='coerce',downcast='integer')


# extracting the first part as category
df['ticket_cat'] = df['Ticket'].apply(lambda s:s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(),np.nan,df['ticket_cat'])
```
