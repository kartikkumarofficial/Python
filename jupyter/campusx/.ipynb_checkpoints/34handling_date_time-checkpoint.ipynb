## Date

```python
# converting into date time object
date['date'] = pd.to_datetime(date['date'], errors='coerce')
```

```python
date['date_year'] = date['date'].dt.year
date.head()
```

```python
date['date_month'] = date['date'].dt.month_name()
# date['date_month'] = date['date'].dt.month
date.head()
```

```python
date['date_day'] = date['date'].dt.day
date['date_day_name'] = date['date'].dt.day_name()

date.head()
```

```python
date['date_isweekend'] = np.where(date['date_day_name'].isin(['Sunday','Saturday']),1,0 )
date.head()
```

```python
# quarter of year
date['date_quarter'] = date['date'].dt.quarter
date.head()
```

```python
# week
date['date_week'] = date['date'].dt.isocalendar().week
date.head()
```

```python
# semester
date['date_semester']=np.where(date['date_quarter'].isin([1,2]),1,2)
date.head()
```

```python
# extract time elapsed between dates

import datetime
today = datetime.datetime.today()
today
```

```python
(today - date['date']).dt.days
```

```python
# months passed

np.round( (today - date['date']) / np.timedelta64(1,'M'),0)
```
