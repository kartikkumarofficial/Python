```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/files/programming/Python/2025/jupyter/datasets/train.csv",usecols=['GarageQual','FireplaceQu','SalePrice'])
```

```python
df['GarageQual'].value_counts().plot(kind='bar')
```

```python
fig = plt.figure() 
ax = fig.add_subplot(111)


df[df['GarageQual']=='TA']['SalePrice'].plot(kind='kde',ax=ax)
df[df['GarageQual'].isnull()]['SalePrice'].plot(kind='kde',ax=ax,color = 'green')

lines , labels = ax.get_legend_handles_labels()
labels = ['Houses with TA','Houses with NA']
ax.legend(lines,labels,loc='best')
plt.title("Garage Quality")
```

```python
temp = df[df['GarageQual']=='TA']['SalePrice']
```

```python
df['GarageQual'].fillna('TA',inplace=True)
```

```python
df['GarageQual'].value_counts().plot(kind='bar')
```

```python
fig = plt.figure()
ax=fig.add_subplot(111)

temp.plot(kind='kde',ax=ax)

df[df['GarageQual'] == 'TA']['SalePrice'].plot(kind='kde', ax=ax, color='red')

lines, labels = ax.get_legend_handles_labels()
labels = ['Original variable', 'Imputed variable']
ax.legend(lines, labels, loc='best')

# add title
plt.title('GarageQual')
```

```python
df['FireplaceQu'].value_counts().plot(kind='bar')
```

```python
df['FireplaceQu'].value_counts().plot(kind='bar')
```

```python
temp = df[df['FireplaceQu']=='Gd']['SalePrice']
```

```python
df['FireplaceQu'].fillna('Gd',inplace=True)
```

```python
df['FireplaceQu'].value_counts().plot(kind='bar')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)


temp.plot(kind='kde', ax=ax)

# distribution of the variable after imputation
df[df['FireplaceQu'] == 'Gd']['SalePrice'].plot(kind='kde', ax=ax, color='red')

lines, labels = ax.get_legend_handles_labels()
labels = ['Original variable', 'Imputed variable']
ax.legend(lines, labels, loc='best')

# add title
plt.title('FireplaceQu')
```

## Using SKLEARN

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['SalePrice']),df['SalePrice'],test_size=0.2)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='most_frequent')
```

```python
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)
```
