```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/files/programming/Python/2025/jupyter/datasets/train.csv",usecols=['Age','Fare','Survived'])
```

```python
df.head()
```

```python
df.isnull().mean()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size = 0.2 ) 
```

```python
X_train['Age_imputed']=X_train['Age']
X_test['Age_imputed']=X_train['Age']
X_train['Fare_imputed']=X_train['Fare']
```

```python
X_train.head()
```

```python
# # chained assignment -> deprecated
# X_train['Age_imputed'][X_train['Age_imputed'].isnull()]=X_train['Age'].dropna().sample(X_train['Age'].isnull().sum(),replace = True).values
# X_test['Age_imputed'][X_test['Age_imputed'].isnull()]=X_train['Age'].dropna().sample(X_test['Age'].isnull().sum(),replace = True).values
```

```python
mask = X_train['Age_imputed'].isnull()

X_train.loc[mask,'Age_imputed']=( X_train['Age'].dropna().sample(mask.sum(),replace=True).values)
```

```python
test_mask=X_test['Age_imputed'].isnull()

X_test.loc[test_mask,'Age_imputed']= (X_train['Age'].dropna().sample(test_mask.sum(),replace=True).values)
```

```python
sns.kdeplot(X_train['Age'],label='Original')
sns.kdeplot(X_train['Age_imputed'],label='Imputed')

plt.legend()
```

```python
print("Original Variance: " , X_train['Age'].var())
print("Imputed Variance: " , X_train['Age_imputed'].var())
```
