```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X,y=make_regression(n_samples=4,n_features=1,n_targets=1,noise=80,random_state=13)
```

```python
X
```

```python
plt.scatter(X,y)
```

```python
 # applying ols->ordinary least square method
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.coef_
# m
```

```python
# b or c
reg.intercept_
```

```python
# les apply grad descent assuming slope is constant m= 78.35 
# les assume the starting value for intercept b=0

y_pred = ((78.35*X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
# now gradually the grad descent->green line will move towards ols (red) line
```

```python
m=78.35
b=0 #start point
loss_slope= -2 * np.sum(y-m*X.ravel() - b)
loss_slope
```

```python
# takin learnig rate as 0.1
lr = 0.1

step_size = loss_slope*lr
step_size
```

```python
# calculating new intercept
b= b - step_size
b
```

```python
y_pred1=((78.35*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
# Iteration 2 
loss_slope = -2 * np.sum(y-m*X.ravel() -b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b= b-step_size
b
```

```python
y_pred2=((78.35*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
# Iteration 3
loss_slope = -2 * np.sum(y-m*X.ravel() -b)
loss_slope
step_size = loss_slope*lr
step_size
b= b-step_size
b
```

```python
y_pred3=((78.35*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred3,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
# doing all this in a loop

b=-100
m=78.35
lr=0.1

epochs=10

for i in range(epochs):
    loss_slope = -2 * np.sum(y- m*X.ravel()-b )
    b=b-(lr*loss_slope)

    y_pred=m*X +b
    plt.plot(X,y_pred)

plt.scatter(X,y)
```

```python
# Creatign custom class to perform grad descent

class GDRegressor:
    def __init__(self,learning_rate,epochs):
        self.m=29.19
        self.b=-120
        self.lr=learning_rate
        self.epochs=epochs
    def fit(self,X,y):
        # calculating b using grad descent
        for i in range(self.epochs):
            loss_slope= -2*np.sum(y-self.m*X.ravel()-b)
            self.b=self.b-(self.lr*loss_slope)
            print(loss_slope,self.b)
        print(self.b)
    
            
            
```

```python
gd=GDRegressor(0.01,100)
```
