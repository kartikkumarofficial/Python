```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import DBSCAN
```

```python
data = load_iris()
X=data.data
y=data.target
```

```python
X.shape
```

```python
y.shape
```

```python
data.feature_names
```

```python
X_plot=X[:,[2,3]] #selectiong only petal length and width
```

```python
# visualizing
plt.figure()
plt.scatter(X_plot[:,0],X_plot[:,1],c=y)
plt.xlabel("petal length")
plt.ylabel('petal width')
plt.title("iris dataset")
```

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.3,min_samples=3)
labels=db.fit_predict(X)
labels
```

```python
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("DBSCAN Clustering (Predicted)")
plt.show()
```
