# Categorical Data

### Bar Chart

```python
print("Total no. of travellers:", df['Survived'].size)
print(df['Survived'].value_counts())
df['Survived'].value_counts().plot(kind='bar')
```

```python
# another way to do the same
import seaborn as sns
sns.countplot(x=df['Survived'],palette=['g','r'])
```

### Pie chart

```python
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```

# Numerical Data

### Histogram

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['Age'],bins=10)
```

### Distplot

```python
sns.displot(df['Age'],kde=True)
# gives probability , skewness as well
```

```python
# another way to find skewness is 
df['Age'].skew()
# 0=perfectly skewed
# 1=positively skewed towards right
# -1=negatively skewed towards left
```
