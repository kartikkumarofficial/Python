```python
import math

num = 16.7

print("Floor:", math.floor(num))
print("Ceil:", math.ceil(num))
print("Square Root:", math.sqrt(num))
print("Integer Square Root:", math.isqrt(25))
print("GCD of 24 and 36:", math.gcd(24, 36))
```

```python
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Array:")
print(arr)

print("Number of Dimensions:", arr.ndim)
print("Shape of Array:", arr.shape)
print("Size of Array:", arr.size)

print("Sum of elements:", arr.sum())
print("Mean of elements:", arr.mean())

print("Sorted Array:")
print(np.sort(arr))

print("Sine of Array Elements:")
print(np.sin(arr))
```

```python
matrix = np.array([[4, 2],
                   [1, 3]])

determinant = np.linalg.det(matrix)
eigenvalues, eigenvectors = np.linalg.eig(matrix)

print("Matrix:")
print(matrix)

print("Determinant:", determinant)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:")
print(eigenvectors)
```

```python
data = [1, 2, 3, 4, 5, 6]

array_1d = np.array(data)

array_2d = array_1d.reshape(2, 3)
array_3d = array_1d.reshape(1, 2, 3)

print("1D Array:", array_1d)
print("2D Array:")
print(array_2d)
print("3D Array:")
print(array_3d)
```

```python
random_matrix_2d = np.random.rand(3, 3)
random_matrix_int = np.random.randint(1, 10, (3, 3))

print("Random Float Matrix:")
print(random_matrix_2d)

print("Random Integer Matrix:")
print(random_matrix_int)
```

```python
from scipy import linalg

matrix = np.array([[2, 1],
                   [5, 3]])

det = linalg.det(matrix)

print("Matrix:")
print(matrix)
print("Determinant using SciPy:", det)
```

```python
matrix = np.array([[4, 2],
                   [1, 3]])

eigen_values, eigen_vectors = linalg.eig(matrix)

print("Matrix:")
print(matrix)

print("Eigen Values:")
print(eigen_values)

print("Eigen Vectors:")
print(eigen_vectors)
```

```python
dataset = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Marks': [85, 90, 78, 92],
    'Age': [20, 21, 19, 22]
}

df = pd.DataFrame(dataset)
print(df)
```

```python
plt.figure()
plt.plot(df['Name'], df['Marks'])
plt.title("Line Plot of Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
plt.figure()
plt.bar(df['Name'], df['Marks'])
plt.title("Bar Plot of Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Age'], df['Marks'])
plt.title("Scatter Plot of Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()

print("Program executed successfully.")
```

```python
from scipy import stats

data = [10, 20, 20, 30, 40, 50]

mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data, keepdims=True)[0][0]
variance = np.var(data)
std_dev = np.std(data)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
data = {
    'Name': ['Amit', 'Neha', 'Rahul', 'Sneha'],
    'Marks': [85, 90, 78, 88]
}

df = pd.DataFrame(data)

filtered_data = df[df['Marks'] > 85]

print("Filtered Data:")
print(filtered_data)
```

```python
data1 = {'ID': [1, 2, 3], 'Name': ['Amit', 'Neha', 'Rahul']}
data2 = {'ID': [1, 2, 3], 'Marks': [85, 90, 78]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

merged_df = pd.merge(df1, df2, on='ID')

print("Merged Dataset:")
print(merged_df)
```

```python
data = {
    'Age': [21, 22, np.nan, 23],
    'Marks': [85, np.nan, 78, 88]
}

df = pd.DataFrame(data)

df_filled = df.fillna(df.mean())

print("Dataset after handling missing values:")
print(df_filled)
```
