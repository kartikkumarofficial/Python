## Missing Values

```python
# capturing all nan values
# first handling categorical features which are missing 

features_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=="O"]
print(len(features_nan))
```

```python
## Let us capture all the nan values
## First lets handle Categorical features which are missing
features_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='O']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
# Replacing missing values with a new label
def replace_cat_feature(dataset,features_nan):
    data=dataset.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data
dataset = replace_cat_feature(dataset,features_nan)
dataset[features_nan].isnull().sum()
```

```python
dataset.head()
```

```python
# Now checking for numerical variablees containing missing values 
numerical_with_nan=[feature for feature in dataset.columns if dataset[feature].dtypes!='object'and dataset[feature].isnull().sum()>1]
# print(len(numerical_with_nan))
for feature in numerical_with_nan:
    print(f"{feature} = {np.round(dataset[feature].isnull().mean(),4)} %") 
```

```python
# replacing the numerical missing value
for feature in numerical_with_nan:
    # we'll replace by using medians since there are outliers
    median_value=dataset[feature].median()
    # create a new feature to capture nan values
    dataset[feature+'_nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
dataset[numerical_with_nan].isnull().sum()
# when doing 'df[col].method(value, inplace=True)', try using 'df.method({col: value}, inplace=True)' or df[col] = df[col].method(value) instead, to perform the operation inplace on the original object
```

```python
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtype!='object']
print(len(numerical_features))
print(numerical_features)
```

```python
temporal_features=[feature for feature in numerical_features if (( 'Yr'  in  feature) or ('Year' in  feature)) and '_nan' not in feature]
print(temporal_features)
```

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    dataset[feature]=dataset['YrSold']-dataset[feature]
```

## Part 2

## Numerical Variables
the numerical features here were skewed so we'll perform log distribution

```python
# taking numerical features that dont have 0 value since we are gonna use log distribution
num_features=[feature for feature in numerical_features if (0 not in dataset[feature].values ) and(feature!='Id')]
print(num_features)
for feature in num_features:
    dataset[feature]=np.log(dataset[feature])
```

```python
dataset.head()
```

## Handling rare categorical feature
removing categorical feature that are present in less than 1% of observations


```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtypes=='object']
print(len(categorical_features))
print(categorical_features)
```

```python
for feature in categorical_features:
    # temp->what fraction of total rows each category represents
    temp = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    # temp_df-w>Filters categories that appear in more than 1% of the dataset , .index extracts the category name
    temp_df = temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'rare_var')

# This avoids categories with very few samples messing up your model
```

## Feature Scaling


```python
feature_scale=[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]
# print(feature_scale)
```

```python
# Standard scaler = will convert data based on standard normal distribution
# MinMaxScaler = will convert data to values b/w 0 and 1
# Standard scaler will also work fine but in this there are no negative values and minmax scaler gave better results

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(dataset[feature_scale]), columns=feature_scale)],
                    axis=1)
```
