```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snb
```

```python
dataset = pd.read_csv("boston_house_prices.csv",header=1)
```

```python
dataset.shape 
```

```python
df=pd.DataFrame(dataset)
df.shape
```

```python
df.head()
```

```python
# Independent and Dependent Variables
X=dataset.drop("MEDV",axis=1) # take all columns of dataset except medv
y=df["MEDV"] # take medv column as dependent variable
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
# standardising the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit_transform(X_train)
```

```python
X_train=scaler.fit_transform(X_train)
# scaler.fit() + scaler.transform() - to directly make changes 
```

```python
X_test = scaler.transform(X_test)
```

```python
# to revert changes
# scaler.inverse_transform(x_train)
```

```python
# implementing linear regression
from sklearn.linear_model import LinearRegression
# cross validation
# Cross-validation is a technique to evaluate your model more reliably by using the dataset multiple times 
# for both training and testing.
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(X,y)
```

```python
mse=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=10)
mse
```

```python
np.mean(mse)
```

```python
# prediction
reg_pred=regression.predict(X_test)
```
